```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/sample_data/car_ownership.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
x=df[["monthly_salary"]]
y=df["owns_car"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
import math
sigmoid = lambda x: 1/(1+math.exp(-x))
```

```python
def logistic(salary,lr):
  x=lr.coef_[0][0]*salary+lr.intercept_[0]
  return sigmoid(x)

```

```python
logistic(62000,lr)
```
